## Funcion para obtener lat y lon a partir de un Geoname

```python
def geonameToLatLon(geoname):
    location = geonames[geonames['geonameid'] == geoname]
    if (len(location['latitude']) == 0 or len(location['longitude']) == 0):
        return np.nan
    return (float(location['latitude']), float(location['longitude']))
    
```

## Obtengo una nueva columna con todas las latitudes y longitudes completas

```python
#reemplazo los NaN de latitudes y longitudes con sus valores reales
latlon = []

properties = properties.reset_index(drop=True)

for i in range(len(properties['lat-lon'])):
    if(properties['lat-lon'][i] is np.nan):
        if(properties['geonames_id'][i] is np.nan):
            latlon.append(np.nan)
        else:    
            latlon.append(geonameToLatLon(properties['geonames_id'][i]))
    else:
        latlon.append(( float(properties['lat'][i]),  float(properties['lon'][i])))

latlon

```
